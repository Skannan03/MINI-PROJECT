#Performing hypothesis testing to find the significant variables
Hypothesis :

H0 (Null Hypothesis) : There is no significant relationship between the variables being tested.

Ha (Alternative Hypothesis) : There is a significant relationship between the variables being tested

Consider significance level as 0.05

```python
from scipy import stats
```

```python
num_cols = df.select_dtypes(include = np.number).columns.to_list()

num_cols.remove('compressive_strength')

print(num_cols)
```

```python
# Num vs Num - f_oneway test

hypo = pd.DataFrame(columns = ['Columns','Pvalue','Remarks'])

for i in num_cols:

    stat , pval = stats.pearsonr(df[i] , df['compressive_strength'])

    hypo= hypo.append({'Columns': i,'Pvalue':pval,'Remarks':'Reject H0' if pval <= 0.05 else 'Failed to reject H0'},ignore_index = True)

hypo
```

Insights from statistical test
From above performed statistical tests we can conclude that all of the columns has rejected null hypotheses which means the columns are significant to the target variable compressive_strength.

```python
performance_df=pd.DataFrame(columns=['Model','R2_train','R2_test','RMSE_train','RMSE_test','Remarks'])
performance_df
```

```python
performance_df = performance_df.append({'Model':'Linear Regression',
                                       'R2_train': r2_score(y_train,preds_train),
                                       'R2_test': r2_score(y_test,preds_test),
                                       'RMSE_train':rmse_train,
                                       'RMSE_test':rmse_test,
                                       'Remarks' : 'Base'} , ignore_index=True)

performance_df
```

```python
def model_performance(model,name):
    global performance_df

    preds_train = model.predict(x_train)
    preds_test = model.predict(x_test)

    rmse_train = np.sqrt(mean_squared_error(y_train,preds_train))
    rmse_test = np.sqrt(mean_squared_error(y_test,preds_test))
    r2_train = r2_score(y_train , preds_train)
    r2_test = r2_score(y_test , preds_test)

    def remark(r2_train,r2_test,rmse_test):
        if abs(r2_train - r2_test) > 0.1 or rmse_test > 7.5:
            return 'Over Fit'

        else:
            return 'Good Fit'

    performance_df = performance_df.append({'Model':name,
                                           'R2_train':r2_train,
                                           'R2_test':r2_test,
                                           'RMSE_train':rmse_train,
                                           'RMSE_test':rmse_test,
                                           'Remarks':remark(r2_train,r2_test,rmse_test)} , ignore_index=True)
```

```python
# Decision tree

from sklearn.tree import DecisionTreeRegressor

model_dt = DecisionTreeRegressor().fit(x_train , y_train)

model_performance(model_dt,'DecisionTree')
```

```python
# KNN

from sklearn.neighbors import KNeighborsRegressor

model_knn = KNeighborsRegressor().fit(x_train , y_train)

model_performance(model_knn , 'KNN')
```

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor().fit(x_train , y_train)

model_performance(model_rf , 'Random Forest')
```

```python
 # AdaBoost

from sklearn.ensemble import AdaBoostRegressor

model_ab = AdaBoostRegressor().fit(x_train , y_train)

model_performance(model_ab , 'AdaBoost')
# Gradient Boosting

from sklearn.ensemble import GradientBoostingRegressor

model_gb = GradientBoostingRegressor().fit(x_train , y_train)

model_performance(model_gb , 'Gradient Boosting')
# Xgboost-RandomForest

from xgboost import XGBRFRegressor

model_xgbrf = XGBRFRegressor().fit(x_train,y_train)

model_performance(model_xgbrf , 'XGBoost Random Forest')
# Neural network
from sklearn.neural_network import MLPRegressor

model_nn = MLPRegressor().fit(x_train , y_train)

model_performance(model_nn , 'Neural Network')
```
